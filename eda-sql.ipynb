### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
%%sql
SELECT DISTINCT Booster_Version
FROM SPACEXTABLE
LIMIT 20;
```

```python
%%sql
SELECT DISTINCT Booster_Version
FROM SPACEXTABLE
WHERE Landing_Outcome = 'Success (drone ship)'
AND PAYLOAD_MASS__KG_ > 4000
AND PAYLOAD_MASS__KG_ < 6000
ORDER BY Booster_Version;
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
# Badly worded question - all missions are sum type of success or failure - so the count of success and failure is the same as the total number of missions.  I am going to assume that the question is asking for the count of missions that were successful and the count of missions that were failures.

%%sql
SELECT Outcome, COUNT(Outcome)
FROM (
    SELECT
        CASE
            WHEN Mission_Outcome LIKE 'Failure%' THEN 'Failure'
            ELSE 'Success'
        END AS Outcome
    FROM SPACEXTABLE
)
GROUP BY Outcome
ORDER BY Outcome DESC;
```

```python
import numpy as np

df["Mission_success"] = np.where(df["Mission_Outcome"].str.startswith("Success"), "Success", "Failure")
df["Mission_success"].value_counts()
```

### Task 8



##### List the names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
# Another badly worded question. The actual booster version is only the first part of the string in the column. The rest of the string is the serial number - which is not the booster version. e.g. F9 B5 B1049.4 - the booster version is F9 B5, the serial number is B1049 and the ".4" - signifies that it is the 4th flight of booster s/n B1049. This would be a much harder question to answer in sqlite if we were to extract the booster version from the string. As it is, from looking at the discussion forum, it seems that the question is asking for the booster version and flight number of the booster with the highest payload mass. This is a very easy question to answer.
```

```python
%%sql
SELECT DISTINCT BOOSTER_VERSION
FROM SPACEXTABLE
WHERE PAYLOAD_MASS__KG_ = (
  SELECT MAX(PAYLOAD_MASS__KG_) FROM SPACEXTBL
)
ORDER BY BOOSTER_VERSION;
```

```python
sorted(df[df["PAYLOAD_MASS__KG_"] == df["PAYLOAD_MASS__KG_"].max()]["Booster_Version"].unique().tolist())
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
%%sql
SELECT 
    substr(Date, 6, 2) AS Month,
    Landing_Outcome,
    Booster_Version,
    Launch_Site
FROM SPACEXTABLE
WHERE substr(Date, 0, 5) = '2015' AND Landing_Outcome = 'Failure (drone ship)'
ORDER BY Date;
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
%%sql
SELECT
    Landing_Outcome,
    COUNT(*) AS COUNT
FROM SPACEXTABLE
WHERE DATE BETWEEN '2010-06-03' AND '2017-03-19'
GROUP BY Landing_Outcome
ORDER BY COUNT DESC;
```
